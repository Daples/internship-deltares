```python
import xarray as xr
import matplotlib.pyplot as plt

mdt_sim = "/home/mverlaan/einf220/fromDavid/saved_results/gtsm/simulation_mdt/gtsm_fine_0000_his.nc"
bias_sim = "/home/mverlaan/einf220/fromDavid/saved_results/gtsm/simulation_bias20/gtsm_fine_0000_his.nc"

mdt_sim_av = "/home/mverlaan/einf220/fromDavid/saved_results/gtsm/simulation_mdt/gtsm_fine_0000_averaged_his.nc"
bias_sim_av = "/home/mverlaan/einf220/fromDavid/saved_results/gtsm/simulation_bias20/gtsm_fine_0000_averaged_his.nc"

mdt_nc = xr.open_dataset(mdt_sim)
bias_nc = xr.open_dataset(bias_sim)

mdt_nc_av = xr.open_dataset(mdt_sim_av)
bias_nc_av = xr.open_dataset(bias_sim_av)
```

```python
station = 27

mdt_nc.isel(stations=station).waterlevel.plot(label="mdt")
bias_nc.isel(stations=station).waterlevel.plot(label="bias20")
plt.legend()
```

```python
station = 30

mdt_nc_av.isel(stations=station).waterlevel.plot(label="mdt")
bias_nc_av.isel(stations=station).waterlevel.plot(label="bias20")
plt.legend()
```
